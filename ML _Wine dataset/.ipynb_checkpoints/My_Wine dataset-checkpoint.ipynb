### EXERCISE 2:
#### Create a scatter plot of the features ash and color_intensity of the wine data set

```python
features= 'ash','color_intensity'

features_index=[data_wine.feature_names.index(features[0]),
                data_wine.feature_names.index(features[1])]
colors=['blue','black','green']
for label, color in zip(range(len(data_wine.target_names)),colors):
    plt.scatter(data_wine.data[data_wine.target==label,features_index[0]],
                data_wine.data[data_wine.target==label,features_index[1]],
               label=data_wine.target_names[label],
                c=color)
    plt.xlabel(features[0])
    plt.ylabel(features[1])
    plt.legend(loc='upper right')
    plt.show
```
