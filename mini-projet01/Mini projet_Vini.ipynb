Affichez les revenus du département "Provence-Alpes-Côte d'Azur " issus de la taxe CVAE des entreprises pour toutes les années (si plusieurs revenus CVAE existent pour une année donnée, vous devez additionner tous les revenus CVAE de cette année).
Vous devez utiliser un DataFrame Pandas.

```python
import pandas as pd
```

```python
doc=pd.read_csv("/home/vini/Desktop/dataset_learning/mini-projet01/dataset.csv")
df=pd.DataFrame(doc)
df
```

```python
df.loc[(df["reg_name"]== "Provence-Alpes-Côte d'Azur") & (df["tax_name"]=="CVAE")].groupby(["year"]).sum()
```

##### Puisqu'il n'y a pas d'année qui se répertent le groupby ne fera pas de regroupement,donc pas de somme

###### Pour donc pouvoir regrouper  tous les revenus CVAE par année, nous ferons ce qui suit.

```python
reg=df.loc[(df["reg_name"]== "Provence-Alpes-Côte d'Azur") & (df["tax_name"]=="CVAE")].groupby("year")["amount"].agg(sum).reset_index()
reg
```
