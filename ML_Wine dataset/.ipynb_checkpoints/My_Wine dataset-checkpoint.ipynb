### EXERCISE 2:
#### Create a scatter plot of the features ash and color_intensity of the wine data set

```python
features= 'ash','color_intensity'

features_index=[data_wine.feature_names.index(features[0]),
                data_wine.feature_names.index(features[1])]
colors=['violet','red','green']
for label, color in zip(range(len(data_wine.target_names)),colors):
    plt.scatter(data_wine.data[data_wine.target==label,features_index[0]],
                data_wine.data[data_wine.target==label,features_index[1]],
               label=data_wine.target_names[label],
                c=color)
    plt.xlabel(features[0])
    plt.ylabel(features[1])
    plt.legend(loc='upper right')
    plt.show
```

### EXERCISE 3:
#### Create a scatter matrix of the features of the wine dataset.

```python
# Scatter matrix of wine data set
data_wine=load_wine()
def rotate_labels(df,axes):
    columns=data_wine.feature_names
    n=len(df.columns)
    for x in range(n):
        for y in range(n):
            # to get the axis of subplots

            ax = axs[x, y]
# to make x axis name vertical
            ax.xaxis.label.set_rotation(90)
# to make y axis name horizontal
            ax.yaxis.label.set_rotation(0)
# to make sure y axis names are outside the plot area
            ax.yaxis.labelpad = 50
data_wine_df = pd.DataFrame(data_wine.data, 
columns=data_wine.feature_names)
axs = pd.plotting.scatter_matrix(data_wine_df,
                                             
c=data_wine.target,
                                             
figsize=(8, 8),
);
rotate_labels(data_wine_df,axs)
```
