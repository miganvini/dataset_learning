```python
X_train.dropna(axis = 0, inplace= True)
```

```python
X_train
```

```python
def pretraitement_text(text) :
    text = text.lower()
    text = text.translate(str.maketrans("", "", string.punctuation)).replace("’"," ")
    
    text = word_tokenize(text)
    stop_word = stopwords.words("english")
    text = [ elt for elt in text if elt not in stop_word]
    text = " ".join(text)
    lemmatizer =  WordNetLemmatizer()
    
    return text
```

```python
X_train["text_nettoyee"] = X_train["Corpus"].apply(pretraitement_text)
```

```python
X_train
```

```python
def generer_wordcloud(phrase):
    tout_texte = ' '.join(phrase)
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(tout_texte)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title("Phrases", fontsize=14)
    plt.axis('off')
    plt.show()
```

```python
sentiment_pos= X_train["text_nettoyee"][X_train.Sentiment=="Positive"]
sentiment_neg=X_train["text_nettoyee"][X_train.Sentiment=="Negative"]
sentiment_neutre=X_train["text_nettoyee"][X_train.Sentiment=="Neutral"]
sentiment_irre=X_train["text_nettoyee"][X_train.Sentiment=="Irrelevant"]
```

```python
generer_wordcloud(sentiment_pos)
```

```python
generer_wordcloud(sentiment_neg)
```

```python
generer_wordcloud(sentiment_neutre)
```

```python
generer_wordcloud(sentiment_irre)
```
