# Exercice 02

Create a function show_transactions that takes a list as parameter and displays each of its elements for example show_transactions ([512,42.08,-12] displays :

You received 512 euros

You received 42.08 euros

You spent 12 euros


```python
show_transactions=[512,42.08,-12,-58]
show_transactions                
```

```python
def show_transactions(num):
    for num in num:                                    
         if num > 0:
            print("You received", num, "euros")
        
         else:
            print("You spent", -num,"euros")
```

```python
show_transactions([512,-12,42.08,-58])
```
