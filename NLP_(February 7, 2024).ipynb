```python
Corpus.lower()
```

```python
Corpus.replace(".",' ').replace(","," ").replace('\n'," ").replace("  "," ").replace("   "," ")
```

```python
def tok_stop_lem(Corpus) :
    Corpus = Corpus.lower()
    Corpus = Corpus.translate(str.maketrans("", "", string.punctuation)).replace("’"," ")
    
    Corpus = word_tokenize(Corpus)
    stop_word = stopwords.words("english")
    Corpus = [ elt for elt in Corpus if elt not in stop_word]
    Corpus = " ".join(Corpus)
    lemmatizer =  WordNetLemmatizer()
    
    return Corpus
```

```python
tok_stop_lem(Corpus)
```

```python
corpus_new=tok_stop_lem(Corpus)
```

```python
def generer_wordcloud(corpus_new):
    wordcloud = WordCloud(width=800, height=400, background_color='Pink').generate(corpus_new)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title("Corpus", fontsize=14)
    plt.axis('off')
    plt.show()
```

```python
generer_wordcloud(Corpus)
```
